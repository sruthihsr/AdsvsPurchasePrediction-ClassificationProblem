```python
import pandas as pd
import numpy as np
```

```python
dataset=pd.read_csv('Social_Network_Ads.csv')
```

```python
print(dataset.shape)
print(dataset.describe())
print(dataset.head(5))
```

```python
iv=dataset.iloc[:,2:4].values
dv= dataset.iloc[:,-1].values
```

```python
print(dv)
```

```python
from sklearn.model_selection import train_test_split
iv_train,iv_test,dv_train,dv_test=train_test_split(iv,dv,test_size=0.2,random_state=0)

```

```python
help(train_test_split)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
iv_train=sc.fit_transform(iv_train)
```

```python
iv_train
```

```python
iv_test=sc.transform(iv_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
DT_Classifier=DecisionTreeClassifier()
DT_Classifier.fit(iv_train,dv_train)
print(DT_Classifier.predict(iv_test))
print(dv_test)
```
